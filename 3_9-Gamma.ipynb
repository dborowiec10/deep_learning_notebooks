```python
import numpy as np
import matplotlib.pyplot as plt

# Gamma distribution:
# # Rarely used on its own - typically a basis for exponential dist, chi-squared dist, erlang dist
# Parameters:
# # a = k   --> shape parameter
# # beta = 1/0 --> rate parameter
```

## Probability Density Function

```python
from scipy.stats import gamma
fig, ax = plt.subplots(1, 2)

a = 40

mean, var, skew, kurt = gamma.stats(a, moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skew: ", skew)
print("Kurtosis: ", kurt)
print("STDDEV: ", gamma.std(a))

x = np.linspace(gamma.ppf(0.01, a), gamma.ppf(0.99, a), 100)
ax[0].plot(x, gamma.pdf(x, a), 'r-', lw=5, alpha=0.6, label='gamma pdf')
ax[1].bar(x, gamma.pdf(x, a))
```

```python
# CDF - probability that x will take value less than or equal to x'
fig, ax = plt.subplots(1, 2) 

ax[0].plot(x, gamma.cdf(x, a), 'bo')
ax[1].bar(x, gamma.cdf(x, a))
plt.show()
```

```python
vals = gamma.rvs(a, size=100000)
fig, ax = plt.subplots(1, 1)
ax.scatter(range(len(vals)), vals)
plt.show() 
```
