```python
import numpy
import matplotlib.pyplot as plt

# Continuous Uniform Distribution

# any interval of numbers of equal width has an equal probability of being observed, the curve is thus a rectangle
# area under the curve must be equal to 1
# length of the interval determines the height of the curve
# Since area under curve needs to be 1, height is set to 1/(b - a)

# describes an experiment where there is an arbitrary outcome that lies between certain bounds
# The bounds are defined by the parameters, a and b, which are the minimum and maximum values
# The difference between the bounds defines the interval length
# All intervals of the same length on the distribution's support are equally probable
```

## Probability Density Function

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)

mean, var, skew, kurt = uniform.stats(moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skew: ", skew)
print("Kurtosis: ", kurt)
print("STDDEV: ", uniform.std())

x = np.linspace(uniform.ppf(0.01), uniform.ppf(0.99), 100)
ax[0].plot(x, uniform.pdf(x), 'r-', lw=5, alpha=0.6, label='uniform pdf')
ax[1].bar(x, uniform.pdf(x))

```

```python
# CDF - probability that x will take value less than or equal to x'
fig, ax = plt.subplots(1, 2) 

ax[0].plot(x, uniform.cdf(x), 'bo')
ax[1].bar(x, uniform.cdf(x))
plt.show()
```

```python
# Plot some random variables according to this distribution
vals = uniform.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
ax.scatter(range(len(vals)), vals)
plt.show()
```
