## Probability Density Function

```python
from scipy.stats import laplace
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)

mean, var, skew, kurt = laplace.stats(moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skew: ", skew)
print("Kurtosis: ", kurt)
print("STDDEV: ", laplace.std())

x = np.linspace(laplace.ppf(0.01), laplace.ppf(0.99), 100)
ax[0].plot(x, laplace(loc=3).pdf(x), 'r-', lw=5, alpha=0.6, label='norm pdf')
ax[1].bar(x, laplace(loc=3).pdf(x))
```

```python
# CDF - probability that x will take value less than or equal to x'
fig, ax = plt.subplots(1, 2) 

ax[0].plot(x, laplace(loc=3).cdf(x), 'bo')
ax[1].bar(x, laplace(loc=3).cdf(x))
plt.show()
```

```python
# Plot some random variables according to this distribution
vals = laplace(loc=3).rvs(size=10000)
fig, ax = plt.subplots(1, 1)
ax.scatter(range(len(vals)), vals)
plt.show()
```
