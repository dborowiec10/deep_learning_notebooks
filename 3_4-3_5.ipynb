```python
# The sum rule:

# Sometimes we know the joint probabilities and need to find the marginal probabilities
# Probabilities written in the margin are sums of joint probabilities for each row and column
```

```python
# x is related to die rolls
# y is related to coin flips

# The meaning of the equation:
# = The probability that x = x is equal to the sum of probabilities corresponding to x = x over every y.
```

```python
# I.e. a probability P(x = 1, y = 'heads') = 1/12
# so... what's the probability of rolling a 1?
# it is the sum of probabilities over y when x = 1. 
# 2 possibilities in this case.
# we can have x = 1 and y = 'heads' or x = 1 and y = 'tails'
# we sum these probabilities to give 1/12 + 1/12 = 1/6
```

```python
# Probabilities for continuous variables:

# same trick but need to use integration instead of summation

# For a continuous variable, probability for the variable to take a "specific" value is close to 0 because there are 
# infinite possible values that the variable can take

# So lets say we have 2 probability density functions separately for x and y
# and we want to calculate marginal probability for x and y

# we need 3 dimensions to represent the probability density function
# 1 for the value of x
# 1 for the value of y
# 1 for the probability associated with getting both of these values

# lets say that both x and y are represented by normal distributions

np.random.seed(123)
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
sns.distplot(x)
plt.title('x')
plt.xlim(-4, 4)
plt.show()
sns.distplot(y)
plt.title('y')
plt.xlim(-4, 4)
plt.show()
```

```python
# in the above plots:
# y axis gives the probability density that the variable takes the value given by x axis
# the area under the curve must be calculated for certain range of the x axis to get the probability to get a value into that range
```

```python
# now... to get marginal probability of x, we need area under the curve of the probability density function along the y-axis.
```

To calculate the marginal probability P(x), we calculate the area under the curve of the projection of the variable on the x-axis.

To calculate the marginal probability P(y), we calculate the area under the curve of the projection of the variable on the y-axis.

```python
# Conditional Probability

# probability of an event happening given another event occured

# Dependent and independent events

# Independent events: dice and coin
# # 2 random variables
# # question: what is the probability to get a 3 from the dice and 'head' from the coin?
# # 2 are independent because we don't expect any effect of one on the other

# Depdendent events: 2 cards from 1 deck
# # probabilities change when we take 1st card and when we take second
# # this makes things depdendent
# # question: what is the probability to draw a 6 on the second drw when we got a 6 on the first draw?

# equation becomes P(y = y | x = x)
# conditional probability
```

```python
# there is a relationship between joint probabilities and conditional probabilities
```

The probability that the random variable y takes the value y given that the random variable x took the value x is the ratio of the probability that both events occur (y takes the value y and x takes the value x) and the probability that x takes the value x.

It can be understood like that: to calculate the probability that both events occur, we have to take the probability that the first event occurs (P(x=x)) and multiply it with the probability that the second event occurs given that the first event occurred (P(y=y|x=x)).

```python
# Multiplying probabilities:
# what is the probability of tossing heads twice?
# 1/2 * 1/2 = 1/4

# i.e. for independent events:
```

```python
# i.e for dependent events
# 2 card draw:
# 1 card drawn = x
# 2 card drawn = y
# we want to know the probability to draw a 6 on the second draw, given we have drawn 6 on the first draw
# P(y = y | x = x)

# so...
# P(y = 6 | x = 6)
# we know that P(y = 6) = 3/51
# we know that P(x = 6) = 4/52
# using the multiplication rule:
# 4/52 * 3/51 = 12/2652 = 1/221

# so the conditional probability is:
# P(y = 6 | x = 6) = P(y = 6, x = 6) / P(x = 6) = (1/221) / (4/52) = 0.0588
```
