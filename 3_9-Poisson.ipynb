```python
import numpy as np
import matplotlib.pyplot as plt

# Poisson Distribution:

# Poisson random variable is typically used to model the numer of times an event happened in a time interval
# I.e. number of users visited on a website in an interval --> Poisson process
# Parameters:
# # mu - rate at which events happen (i.e. event can occur 0, 1, 2 ... times in an interval)
# avg number of events in interval -> lambda (event rate, rate parameter)
# probability of observing k events in an interval is given by:
```

## Probability Mass Function

```python
# Normal (Gaussian) Distribution is a limiting case of Poisson distribution with lambda --> infinity
# I times between random events follow an exponential distribution with rate lambda, 
# the total number of events in time period of length t follows the Poisson distribution with parameter t(lambda)
```

```python
from scipy.stats import poisson
fig, ax = plt.subplots(1, 2)

mu = 3

mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skew: ", skew)
print("Kurtosis: ", kurt)
print("STDDEV: ", poisson.std(mu))

x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
ax[0].plot(x, poisson.pmf(x, mu), 'r-', lw=5, alpha=0.6, label='norm pdf')
ax[1].bar(x, poisson.pmf(x, mu))

```

```python
# CDF - probability that x will take value less than or equal to x'
fig, ax = plt.subplots(1, 2) 

ax[0].plot(x, poisson.cdf(x, mu), 'bo')
ax[1].bar(x, poisson.cdf(x, mu))
plt.show()
```

```python
# Plot some random variables according to this distribution
vals = poisson.rvs(mu, size=10)
fig, ax = plt.subplots(1, 1)
ax.scatter(range(len(vals)), vals)
plt.show()
```
