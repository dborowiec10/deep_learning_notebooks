```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
# Norms of vectors:

# Functions with following characteristics
# 1. non-negative values - think of them as lenght
# 2. only 0 if vector is 0
# 3. norms respect triangle inequity
# 4. || k . u || = | k | . | u | = the norm of vector multiplied by scalar is equal to absolute value of scalar multiplied by norm of vector
# # typically expressed by ||x|| where x is the vector

# triangle inequity:
# norm of sum of some vectors is less than or equal to sum of the norms of those vectors
# || u + v || <= ||u|| + ||v||

# Example:
```

```python
u = np.array([1, 6])
v = np.array([4, 2])
```

```python
# manually calculating norm of 2 vectors:
uv_norm = math.sqrt(  ((u[0] + v[0])**2) + ((u[1] + v[1])**2))
```

```python
u_norm = math.sqrt((u[0]**2) + (u[1]**2))
```

```python
v_norm = math.sqrt((v[0]**2) + (v[1]**2))
```

```python
u_v_norm = u_norm + v_norm
```

```python
print("v_norm: ", v_norm)
print("u_norm: ", u_norm)
print("uv_norm: ", uv_norm)
print("u_v_norm: ", u_v_norm)
```

```python
# so the norm of sum of vectors (uv_norm) is < norms of vectors summed (u_v_norm)
```

```python
# can also do the same in numpy

print("uv_norm: ", np.linalg.norm(u+v))
print("u_v_norm: ", np.linalg.norm(u) + np.linalg.norm(v))
```

```python
# P-norms of vectors = recipe:

# Calculate abs of each element
# take power p of each element
# sum all powered values
# take power of 1/p of the sum

# ||x||p = (Sum_i |xi|p)^1/p
```

```python
# L0 norms:
# correspond to number of non-0 elements in vector = each element raised to power of 0 except 0 will give 1
```

```python
# L1 norms:
# essentially a sum of all absolute values of the vector
```

```python
# Squared L2 norm:
# convenient - removes square root
# can be easily calculated using xT.x = i.e. Trasnspose of x dot-product x
```

```python
# i.e.

x = np.array([
    [2],
    [5],
    [3],
    [3]
])

sq_eq_norm = x.T.dot(x)
sq_eq_norm_numpy = np.linalg.norm(x)**2
```

```python
print(sq_eq_norm)
print(sq_eq_norm_numpy)
```

```python
# dot product of vectors can be re-written using norms
```

```python


# for vectors:
# x = [0]
#     [2]

# and

# y = [2]
#     [2]

x = [0,0,0,2]
y = [0,0,2,2]

plt.xlim(-2, 4)
plt.ylim(-2, 5)
plt.axvline(x=0, color='grey', zorder=0)
plt.axhline(y=0, color='grey', zorder=0)

plt.quiver([x[0], y[0]],
           [x[1], y[1]],
           [x[2], y[2]],
           [x[3], y[3]],
           angles='xy', scale_units='xy', scale=1)

plt.text(-0.5, 1, r'$\vec{x}$', size=18)
plt.text(1.5, 0.5, r'$\vec{y}$', size=18)

# the angle 0 is 45 degrees


x = np.array([0, 2])
y = np.array([2, 2])

print("Regular dot product: ", x.T.dot(y))
print("Using norms: ", (np.linalg.norm(x) * np.linalg.norm(y)) * np.cos(np.deg2rad(45)))
```
