```python
# Bernoulli Distribution:
# # Conditions:
# # # There must be 1 trial
# # # There must be only 2 possible outcomes of the trial (success, failure)
# # # P(Success) = p
# # # P(Failure) = 1 - p = q
# # # Convention: assign value 1 to event with probability p and 0 to probability q
# # # Convention: p > 1 - p
# # # There must be available the probability of 1 of the events (Success/Failure) or some past data indicating experimental probability
# # # If we satisfy all these conditions, we can say that discrete random variable X follows Bernoulli distribution with probability of success of p
# # # :: X ~ Bern(p)
```

## Probability Mass Function

```python
from scipy.stats import bernoulli
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)

# probability
p = 0.3
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skew: ", skew)
print("Kurtosis: ", kurt)
print("STDDEV: ", bernoulli.std(p))

# calculate inverse cdf for bernoulli using percent point function (ppf) to get x value(s)
ppf_lower = bernoulli.ppf(0.01, p)
ppf_upper = bernoulli.ppf(0.99, p)
x = np.array([ppf_lower, ppf_upper])

# plot probability mass function
ax[0].plot(x, bernoulli.pmf(x, p), 'bo', ms=8)
ax[1].bar(x, bernoulli.pmf(x, p))

plt.show()
```

```python
# CDF - probability that x will take value less than or equal to x'
fig, ax = plt.subplots(1, 2) 
p = 0.3
ppf_lower = bernoulli.ppf(0.01, p)
ppf_upper = bernoulli.ppf(0.99, p)
x = np.array([ppf_lower, ppf_upper])
ax[0].plot(x, bernoulli.cdf(x, p), 'bo')
ax[1].bar(x, bernoulli.cdf(x, p))
plt.show()
```

```python
# Plot some random variables according to this distribution
p = 0.3
vals = bernoulli.rvs(p, size=10)
fig, ax = plt.subplots(1, 1)
ax.scatter(range(len(vals)), vals)
plt.show()
```
