```python
# The logistic sigmoid is commonly used to produce the φ parameter of a Bernoulli distribution because its range is (0,1), which lies within the valid range of values for the φ parameter.
# The sigmoid function saturates when its argument is very positive or very negative, meaning that the function becomes very flat and insensitive to small changes in its input.

import numpy as np
import matplotlib.pyplot as plt
import math

def logistic_sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10, 10)

fig, ax = plt.subplots(1, 1)
ax.plot(x, logistic_sigmoid(x), 'bo', ms=8)
plt.show()
```

```python
# The softplus function:
# # The softplus function can be useful for producing the β or σ parameter of a normal distribution because its range is (0,∞).
# # It also arises commonly when manipulating expressions involving sigmoids.
```

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, softplus(x), 'bo', ms=8)
plt.show()
```
