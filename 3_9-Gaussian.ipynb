The two parameters µ ∈ R and σ ∈ (0,∞) control the normal distribution.

The parameter µ gives the coordinate of the central peak. 

This is also the mean of the distribution: E[x] = µ. 

The standard deviation of the distribution is given by σ, and the variance by σ^2.

```python
# µ gives the mean/central coordinate
# σ^2 = stddev of distribution
# σ = variance of distribution
# β = precision parameter (inverse variance of the distribution)
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)

mean, var, skew, kurt = norm.stats(moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skew: ", skew)
print("Kurtosis: ", kurt)
print("STDDEV: ", norm.std())

x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
ax[0].plot(x, norm.pdf(x), 'r-', lw=5, alpha=0.6, label='norm pdf')
ax[1].bar(x, norm.pdf(x))
```

```python
# CDF - probability that x will take value less than or equal to x'
fig, ax = plt.subplots(1, 2) 

ax[0].plot(x, norm.cdf(x), 'bo')
ax[1].bar(x, norm.cdf(x))
plt.show()
```

```python
# Plot some random variables according to this distribution
vals = norm.rvs(size=10000)
fig, ax = plt.subplots(1, 1)
ax.scatter(range(len(vals)), vals)
plt.show()
```
