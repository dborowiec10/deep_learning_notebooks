```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagonal matrices
```

```python
# square diagonal matrix
A = np.array([
    [2, 0, 0, 0],
    [0, 4, 0, 0],
    [0, 0, 3, 0],
    [0, 0, 0, 1]
])

# non-square diagonal matrix
B = np.array([
    [2, 0, 0],
    [0, 4, 0],
    [0, 0, 3],
    [0, 0, 0]
])

# also non-square diagonal matrix
C = np.array([
    [2, 0, 0, 0],
    [0, 4, 0, 0],
    [0, 0, 3, 0]
])

# generally represented by diag(v) where v is the vector containing diagonal values.
# in numpy, can be done with:

print(np.diag(np.array([2, 3, 3, 1])))
```

```python
# multiplying diagnoal matrices is thus simple element multiplication of diagonal vector with another vector

print(A*np.array([3, 3, 3, 3]))
```

```python
# inverses of diagonal matrices are easy to find under some conditions:
# all diagonal entries must be non-0
# matrix must be square

# i.e.:
A_inv = np.linalg.inv(A)
```

```python
# to check, dot-product A-1 with A to get identity matrix
A_inv.dot(A)
```

```python
# Symmetric matrices:
# they're equal to their own transpose. I.e. AT = A

A = np.array([[2, 4, -1], [4, -8, 0], [-1, 0, 3]])
AT = A.T
print(A)
print(AT)
```

```python
# Orthogonal vectors:
# vectors separated by 90deg angle in cartesian space
# dot product gives 0
```

```python
x = [0,0,2,2]
y = [0,0,2,-2]

plt.quiver([x[0], y[0]],
           [x[1], y[1]],
           [x[2], y[2]],
           [x[3], y[3]],
           angles='xy', scale_units='xy', scale=1)

plt.xlim(-2, 4)
plt.ylim(-3, 3)
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.text(1, 1.5, r'$\vec{u}$', size=18)
plt.text(1.5, -1, r'$\vec{v}$', size=18)
plt.show()
```

```python
x = np.array([2, 2])
y = np.array([2, -2])

print(x.T.dot(y))
```

If both
vectors have nonzero norm, this means that they are at a 90 degree angle to each
other.

In Rn , at most n vectors may be mutually orthogonal with nonzero norm.

If the vectors are not only orthogonal but also have unit norm, we call them
*orthonormal*.

It is impossible to have more than n vectors mutually orthogonal in Rn. 

For instance try to draw 3 vectors in a 2-dimensional space (R2) that are mutually orthogonal…

```python
# Orthogonal matrices:
# A matrix is orthogonal if columns are mutually orthogonal and have a unit norm (orthonormal) and rows are mutually orthonormal and have unit norm.
```

```python
# orthogonal matrices have interesting properties

# ATA = AAT = I
```

```python
A = np.array([
    [np.cos(50), -np.sin(50)], 
    [np.sin(50), np.cos(50)]
])
```

```python
col0 = A[:, 0].reshape(A[:, 0].shape[0], 1)
col1 = A[:, 1].reshape(A[:, 1].shape[0], 1)

row0 = A[0, :].reshape(A[0, :].shape[0], 1)
row1 = A[1, :].reshape(A[1, :].shape[0], 1)

# columns are orthogonal
print(col0.T.dot(col1))

# rows are orthogonal
print(row0.T.dot(row1))

# ATA should give identity matrix
print(A.T.dot(A))
print("---------------")
print(A.T)
print(np.linalg.inv(A))
```
