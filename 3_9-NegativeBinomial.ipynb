```python
import numpy as np
import matplotlib.pyplot as plt

# Negative Binomial Distribution:
# 2 outcomes are possible (success, failure)
# when probability of success and failure are the same for all trials --> we observe Binomial Distribution
# outcomes need not be equally likely and each trial is independent from each other
```

Negative Binomial Distribution models the number of successes in a sequence of independent and identically distributed Bernoulli trials before a specified (non-random) number of failures (denoted r) occurs.

For example, we can define rolling a 6 on a die as a failure, and rolling any other number as a success, and ask how many successful rolls will occur before we see the third failure (r = 3).

## Probability Mass Function

```python
from scipy.stats import nbinom
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)

r = 3 # number of successes
p = 0.3 # probability of single success

mean, var, skew, kurt = nbinom.stats(r, p, moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skew: ", skew)
print("Kurtosis: ", kurt)
print("STDDEV: ", nbinom.std(r, p))

# calculate inverse cdf for binom using percent point function (ppf) to get x value(s)
ppf_lower = nbinom.ppf(0.01, r, p)
ppf_upper = nbinom.ppf(0.99, r, p)
x = np.arange(ppf_lower, ppf_upper)

# plot probability mass function
ax[0].plot(x, nbinom.pmf(x, r, p), 'bo', ms=8)
ax[1].bar(x, nbinom.pmf(x, r, p))

plt.show()

# there is 0.3 chance that first failure is after x = 0, 1, 2, 3...
```

```python
# CDF - probability that x will take value less than or equal to x'
fig, ax = plt.subplots(1, 2) 

ax[0].plot(x, nbinom.cdf(x, r, p), 'bo')
ax[1].bar(x, nbinom.cdf(x, r, p))
plt.show()
```

```python
# Plot some random variables according to this distribution
vals = nbinom.rvs(r, p, size=10)
fig, ax = plt.subplots(1, 1)
ax.scatter(range(len(vals)), vals)
plt.show()
```
