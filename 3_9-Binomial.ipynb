```python
import numpy
import matplotlib.pyplot as plt

# Binomial Distribution:
# 2 outcomes are possible (success, failure)
# when probability of success and failure are the same for all trials --> we observe Binomial Distribution
# outcomes need not be equally likely and each trial is independent from each other
# # Parameters:
# # # n --> total number of trials
# # # p --> probability of success in each trial
# Bernoulli distribution is a special case of binomial distribution with only 1 trial!
```

## Probability Mass Function

```python
# Can be understood as:
# n = n independent Bernoulli trials
# k successes occur with probability p^k and n-k failures occur with probability (1 - p)^(n - k)
# However, the k successes can occur anywhere among the n trials and there are (n/k) different ways of distributing k successes in a sequence of n trials
```

```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)

# probability
n = 10 # 1000 trials
p = 0.5 # probability of success

mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skew: ", skew)
print("Kurtosis: ", kurt)
print("STDDEV: ", binom.std(n, p))

# calculate inverse cdf for binom using percent point function (ppf) to get x value(s)
ppf_lower = binom.ppf(0.01, n, p)
ppf_upper = binom.ppf(0.99, n, p)
x = np.arange(ppf_lower, ppf_upper)

# plot probability mass function
ax[0].plot(x, binom.pmf(x, n, p), 'bo', ms=8)
ax[1].bar(x, binom.pmf(x, n, p))

plt.show()
```

```python
# CDF - probability that x will take value less than or equal to x'
fig, ax = plt.subplots(1, 2) 

ax[0].plot(x, binom.cdf(x, n, p), 'bo')
ax[1].bar(x, binom.cdf(x, n, p))
plt.show()
```

```python
# Plot some random variables according to this distribution
vals = binom.rvs(n, p, size=10)
fig, ax = plt.subplots(1, 1)
ax.scatter(range(len(vals)), vals)
plt.show()
```
